```python
import pandas as pd
#fb=pd.read_table("Restaurant_Reviews.tsv")
#bf=pd.read_table("Restaurant_Reviews.tsv",header=None)154849joij
fb=pd.read_table("FB_HI_EN_CR.tsv",header=None,names=['word','sense'])
fb.head()
```

```python
# filter the DataFrame using an OR condition
word_hi_en = fb[(fb.sense=='hi') | (fb.sense=='en')]
```

```python
word_hi_en.sense.value_counts()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.cross_validation import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
word_hi_en .loc[word_hi_en ["sense"]=='hi',"sense",]=1
word_hi_en .loc[word_hi_en ["sense"]=='en',"sense",]=0
```

```python
word_hi_en .head()
```

```python

X=word_hi_en.word
y=word_hi_en.sense
```

```python
# split X and y into training and testing sets
from sklearn.cross_validation import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
# examine the object shapes
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# import and instantiate CountVectorizer
from sklearn.feature_extraction.text import CountVectorizer
vect = CountVectorizer()
```

```python
# fit and transform X_train into X_train_dtm
X_train_dtm = vect.fit_transform(X_train)
X_train_dtm.shape
```

```python
# transform X_test into X_test_dtm
X_test_dtm = vect.transform(X_test)
X_test_dtm.shape
```

```python
# import and instantiate MultinomialNB
from sklearn.naive_bayes import MultinomialNB
nb = MultinomialNB()
```

```python
# train the model using X_train_dtm
#nb.fit(X_train_dtm, y_train)
%time nb.fit(X_train_dtm, y_train.astype(str))
```

```python
y_pred_class = nb.predict(X_test_dtm)
```

```python
from sklearn import metrics
metrics.accuracy_score(y_test.astype(str), y_pred_class.astype(str))
```

```python
metrics.confusion_matrix(y_test.astype(str), y_pred_class.astype(str))
```
